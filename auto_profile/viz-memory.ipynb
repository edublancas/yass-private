```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (18, 5)
```

```python
def load(filename):
    df = pd.read_csv(filename, sep=' ',
                 names=['mem', 'usage', 'time'],
                skiprows=[0])
    return df
```

```python
def plot(filename):
    df = load(filename)
    plt.plot(df.time, df.usage)
```

```python
# old
plot('mprofile_20180301002826.dat')
```
